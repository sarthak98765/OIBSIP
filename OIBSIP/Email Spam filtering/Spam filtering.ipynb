```python
from nltk.corpus import stopwords

stopwords.words('english')
```

```python
import string

string.punctuation
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

ps.stem('writting')
```

```python
def transform_text(text):
    
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    
    for i in text:
        if i.isalnum():
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
transform_text('This is SAAD ALI here, and we the partners done the Machine Learning project1267, what About you? idno: ##$#$%%')
```

# Model Building

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer(max_features=3000)
X = tfid.fit_transform(df['transformed_text']).toarray()
X
```

```python
X.shape
```

```python
y = df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.naive_bayes import BernoulliNB,GaussianNB,MultinomialNB
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
```

```python
Ber = BernoulliNB()
Gau = GaussianNB()
Mul = MultinomialNB()
```

# BernouliNB() 

```python
Ber.fit(X_train,y_train)
```

```python
y_pred_Ber = Ber.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred_Ber))
print(confusion_matrix(y_test,y_pred_Ber))
print(precision_score(y_test,y_pred_Ber))
```

# GaussianNB()

```python
Gau.fit(X_train,y_train)
```

```python
y_pred_Gau = Gau.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred_Gau))
print(confusion_matrix(y_test,y_pred_Gau))
print(precision_score(y_test,y_pred_Gau))
```
